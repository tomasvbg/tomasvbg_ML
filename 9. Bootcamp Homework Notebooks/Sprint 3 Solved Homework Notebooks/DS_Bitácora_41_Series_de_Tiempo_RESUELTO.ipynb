### Personal Notes Time Series

`Series de tiempo:` 
- variable medida secuencialmente en el tiempo.
- usos:
    - clasificacin/regresin
    - forecasting
    - deteccin de puntos de quiebre

***  

***Tipos de Series de tiempo***

`- Estacional:` representan ciclos (el valor del punto de arranque es similar el punto final)

`- Estacionaria:` orbitan en torno a un valor medio -no tiene tendencia-. 

- Una serie puede ser estacionaria y estacional al mismo tiempo.

***

***Conceptos relevantes***

`- Frecuencia de muestreo:` diferencia entre dos tiempos consecutivos (distancia). Se espera que las medidas de comparacin de una serie de tiempo sean equidistantes.

`- Tendencia:` comportamiento a largo plazo de la serie (crece, decrece, o estacionaria en la media -horizontal sin inclinacin-).

`- Estacionalidad:`existen ciertos errores sobre la serie que se repiten regularmente (alrededor de la misma fecha, por ejemplo).

`- Ruido y anomalas:` lo que no  entra en los elementos anteriores. Si tengo mucho error, significa que mi modelo no ha capturado la escencia de los datos -underfitting- (necesito ms preprocesamiento o ms datos). 
    
   - `ruido blanco (deseable)`: el error no tiene un comportamiento predecible (no se puede modelar/ no se puede detectar un patrn).
    
   -`caminata al azar (el model no est absorviendo una componente, se va al error)`: el error tiene un comportamiento predecible (se puede modelar).
    
- Existen varias formas de modelar cmo se combinan estos componentes.

`- Media movil`
El proceso de tomar una ventana, trasladarlo a lo largo de una serie de tiempo y hacer alguna operacin en esa ventana se conoce como *rolling window*. En general, se utiliza junto con la media, pero tambin se podra obtener el mximo, mnimo o cualquier otro estadstico.

## 2. Componentes de la Serie

Obtener componentes de una Serie de Tiempo es un tema amplio y, como siempre, depende del problema en el que ests trabajando. Algunas series se comportan como $Tendencia + Estacionalidad + Ruido$, mientras que otras tienen comportamientos ms complejos, como $Tendencia * Estacionalidad + Ruido$. Existen muchas formas de obtener las componentes, pero no todas funcionarn en todos los casos. Adems, cada tcnica tiene sus ventajas y desventajas. Vamos a hacer una enumeracin y demostracin no exhaustiva de algunos mtodos, simplemente para que te lleves una idea de qu cosas se pueden hacer fcilmente.

* El mtodo ms sencillo se conoce como *media mvil*. Consiste en tomar una ventana de tiempo suficientemente grande y promediar la serie sobre esa ventana. De esta forma, obtienes la tendencia. Tomate un tiempo para pensar qu significa *suficientemente grande*. Una caracterstica de este mtodo es que no asume ninguna forma para la serie de tiempo - a diferencia del que te contaremos a continuacin - as que suele funcionar en muchos casos.

* Supongamos que tienes motivos para creer que la tendencia sigue un comportamiento lineal. Entonces, basta con ajustar una regresin lineal a la serie (antes tambin la puedes haber promediado con el mtodo anterior) para obtener la tendencia. Este mtodo tambin sirve si tienes motivos para creer que la tendencia sigue una forma arbitraria pero conocida (por ejemplo, cuadrtica, exponencial, tipo seno o coseno, etc.).

Una vez que obtienes la tendencia, en ambos casos puedes tomar la serie original, restar la tendencia y obtener una nueva serie sin tendencia, donde esperamos que sea la componente estacional la que domine.

Luego, para obtener componentes estacionales, se suele asumir una forma armnica (tipo seno o coseno). Es decir, se puede hacer un procedimiento similar al que hiciste con la tendencia (ajustar y restar a la serie sin tendencia) para obtener ruido + componentes restantes.

Veamos un poco como se traduce esto en el cdigo.

**Media movil**

El proceso de tomar una ventana, trasladarlo a lo largo de una serie de tiempo y hacer alguna operacin en esa ventana se conoce como *rolling window*. En general, se utiliza junto con la media, pero tambin se podra obtener el mximo, mnimo o cualquier otro estadstico.

Vamos a operar sobre la serie semanal promedio. Comenzamos sacando outliers y rellenando valores faltantes. Qu hace la funcin `interpolate`?

```python
# I define that a value over 50 should be trated as an NaN 
semanal.Tmax[semanal.Tmax > 50] = np.nan
semanal.Tmin[semanal.Tmin > 50] = np.nan
```

```python
# I interpolate in order to fill the NaN values
print(semanal.isna().sum())
semanal = semanal.interpolate()
print(semanal.isna().sum())
```

```python
# I create a new column with the mean temperature per week
semanal['Tmean'] = (semanal.Tmax + semanal.Tmin)/2
semanal
```

```python
# I define the size of the window (two years) in order to model the mobile mean
ventana = int(6)
semanal_promedio_tendencia = semanal.Tmean.rolling(window=ventana, center = True).mean()
semanal_promedio_tendencia
```

Por qu aparecen tanto valores faltantes?Cuntos son?A qu corresponde?

Grafiquemos

```python
# I can see that the mobile mean model doesnt adjust to the orginal data tendency
# The origin of this might be the size of the window
plt.figure(figsize = (15,5))
plt.plot(semanal.index, semanal.Tmean, label = 'Temperatura Promedio')
plt.plot(semanal_promedio_tendencia, lw = 4, label = 'Temperatura Promedio - Rolling Window')
plt.xlabel('Fecha')
plt.ylabel('Temperatura (C)')
plt.legend()
plt.show()
```

**Tendencia Lineal**

Este mtodo consiste simplemente en ajustar una regresin lineal a la serie. Ten en cuenta que la validez de este mtodo es muy limitada.

Para comenzar, lo nico que tenemos que tener en cuenta es que no podemos ajustar una regresin lineal con las fechas as como las tenemos, por lo que tenemos que pasarlas a un formato numrico.

```python
semanal['fecha_numeros']=semanal.index.map(datetime.datetime.toordinal)
semanal.head()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(semanal.fecha_numeros.values.reshape(-1, 1), semanal.Tmean)
```

```python
plt.figure(figsize = (15,5))
plt.plot(semanal.index, semanal.Tmean, label = 'Temperatura Promedio')
plt.plot(semanal_promedio_tendencia, ls = '--', lw = 4, label = 'Temperatura Promedio - Rolling Window')
plt.plot(semanal.index, lr.predict(semanal.fecha_numeros.values.reshape(-1, 1)), lw = 4, label = 'Temperatura Promedio - Ajuste Lineal')
# plt.xlim(datetime.date(1985,1,1), datetime.date(2018,1,1))
# plt.ylim([0,30])
plt.xlabel('Fecha')
plt.ylabel('Temperatura (C)')
plt.legend()
plt.show()
```

Qu conclusiones sacaras de ese grfico?

Falta obtener las restantes componentes. 

```python
plt.figure(figsize = (12,9))

plt.subplot(3,1,1)
plt.title('Serie Original')
plt.plot(semanal.index, semanal.Tmean)
plt.ylabel('Temperatura (C)')

plt.subplot(3,1,2)
plt.title('Tendencia Rolling Window')
plt.plot(semanal_promedio_tendencia)
plt.ylabel('Temperatura (C)')

plt.subplot(3,1,3)
plt.title('Diferencia entre la serie original y la tendencia')
plt.plot(semanal.index, semanal.Tmean - semanal_promedio_tendencia)
plt.xlabel('Fecha')
plt.ylabel('Temperatura (C)')

plt.tight_layout()
plt.show()
```

Una forma comn de interpretar las componentes es parndose en la tendencia y diferenciar el aporte de cada estacin. Por ejemplo, podramos decir que la temperatura promedio de Buenos Aires es de alrededor de 17 grados centgrados, con una componente estacional del orden de -7 grados en invierno y del orden de +7 grados en verano.
