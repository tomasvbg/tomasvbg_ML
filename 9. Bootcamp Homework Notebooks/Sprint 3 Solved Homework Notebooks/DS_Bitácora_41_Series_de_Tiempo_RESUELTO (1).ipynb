### Personal Notes Time Series

`Series de tiempo:` 
- variable medida secuencialmente en el tiempo.
- usos:
    - clasificación/regresión
    - forecasting
    - detección de puntos de quiebre

***  

***Tipos de Series de tiempo***

`- Estacional:` representan ciclos (el valor del punto de arranque es similar el punto final)

`- Estacionaria:` orbitan en torno a un valor medio -no tiene tendencia-. 

- Una serie puede ser estacionaria y estacional al mismo tiempo.

***

***Conceptos relevantes***

`- Frecuencia de muestreo:` diferencia entre dos tiempos consecutivos (distancia). Se espera que las medidas de comparación de una serie de tiempo sean equidistantes.

`- Tendencia:` comportamiento a largo plazo de la serie (crece, decrece, o estacionaria en la media -horizontal sin inclinación-).

`- Estacionalidad:`existen ciertos errores sobre la serie que se repiten regularmente (alrededor de la misma fecha, por ejemplo).

`- Ruido y anomalías:` lo que no  entra en los elementos anteriores. Si tengo mucho error, significa que mi modelo no ha capturado la escencia de los datos -underfitting- (necesito más preprocesamiento o más datos). 
    
   - `ruido blanco (deseable)`: el error no tiene un comportamiento predecible (no se puede modelar/ no se puede detectar un patrón).
    
   -`caminata al azar (el model no está absorviendo una componente, se va al error)`: el error tiene un comportamiento predecible (se puede modelar).
    
- Existen varias formas de modelar cómo se combinan estos componentes.

`- Media movil`
El proceso de tomar una ventana, trasladarlo a lo largo de una serie de tiempo y hacer alguna operación en esa ventana se conoce como *rolling window*. En general, se utiliza junto con la media, pero también se podría obtener el máximo, mínimo o cualquier otro estadístico.

## 2. Componentes de la Serie

Obtener componentes de una Serie de Tiempo es un tema amplio y, como siempre, depende del problema en el que estás trabajando. 

`Algunas series se comportan como $Tendencia + Estacionalidad + Ruido$, mientras que otras tienen comportamientos más complejos, como Tendencia * Estacionalidad + Ruido.` 

Existen muchas formas de obtener las componentes, pero no todas funcionarán en todos los casos. Además, cada técnica tiene sus ventajas y desventajas. Vamos a hacer una enumeración y demostración no exhaustiva de algunos métodos, simplemente para que te lleves una idea de qué cosas se pueden hacer fácilmente.

`* El método más sencillo se conoce como *media móvil*.` Consiste en tomar una ventana de tiempo suficientemente grande y promediar la serie sobre esa ventana. `De esta forma, obtienes la tendencia.` Tomate un tiempo para pensar qué significa *suficientemente grande*. Una característica de este método es que no asume ninguna forma para la serie de tiempo - a diferencia del que te contaremos a continuación - así que suele funcionar en muchos casos.

`* Supongamos que tienes motivos para creer que la tendencia sigue un comportamiento lineal.` `Entonces, basta con ajustar una regresión lineal a la serie `(antes también la puedes haber promediado con el método anterior) para obtener la tendencia. `Este método también sirve si tienes motivos para creer que la tendencia sigue una forma arbitraria pero conocida (por ejemplo, cuadrática, exponencial, tipo seno o coseno, etc.).`

`Una vez que obtienes la tendencia, en ambos casos puedes tomar la serie original, restar la tendencia y obtener una nueva serie sin tendencia`, donde `esperamos que sea la componente estacional la que domine.`

`Luego, para obtener componentes estacionales, se suele asumir una forma armónica (tipo seno o coseno).` 

Es decir, `se puede hacer un procedimiento similar` al que hiciste con la tendencia (ajustar y restar a la serie sin tendencia) `para obtener ruido + componentes restantes.``

Veamos un poco como se traduce esto en el código.

**Media movil**

El proceso de tomar una ventana, trasladarlo a lo largo de una serie de tiempo y hacer alguna operación en esa ventana se conoce como `*rolling window*. En general, se utiliza junto con la media, pero también se podría obtener el máximo, mínimo o cualquier otro estadístico.``

Vamos a operar sobre la serie semanal promedio. Comenzamos sacando outliers y rellenando valores faltantes. ¿Qué hace la función `interpolate`?

```python
# I eliminate the values over and below 50 (outliers)
semanal.Tmax[semanal.Tmax > 50] = np.nan
semanal.Tmin[semanal.Tmin > 50] = np.nan
```

```python
# the I see how many outliers where in the dataset
print(semanal.isna().sum())
semanal = semanal.interpolate()
print(semanal.isna().sum())
```

```python
# I create a new column with the temperature mean per week
semanal['Tmean'] = (semanal.Tmax + semanal.Tmin)/2
semanal
```

```python
# I create the window
ventana = int(365*2/7)
semanal_promedio_tendencia = semanal.Tmean.rolling(window=ventana, center = True).mean()
semanal_promedio_tendencia
```

¿Por qué aparecen tanto valores faltantes?¿Cuántos son?¿A qué corresponde?

Grafiquemos

```python
plt.figure(figsize = (15,5))
plt.plot(semanal.index, semanal.Tmean, label = 'Temperatura Promedio')
plt.plot(semanal_promedio_tendencia, lw = 4, label = 'Temperatura Promedio - Rolling Window')
plt.xlabel('Fecha')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.show()
```

### Correlación en SDT
Dos series de tiempo están correlacionadas si su comportamiento en el tiempo es parecido (crecen y decrecen juntos); y  anticorrelacionadas si si el valor en una es máxima y en otro mínimo.



### Autocorrelación en SDT

- Correlación de una serie consigomisma. 

`- Permite determinar el tamaño del lookback (cuántas columnas vamos a crear en el dataset).`

- La correlación de una serie consigomisa es 1.

- Es por eso que calculamos la correlación entre la serie original y la misma serie con lags de tiempo.


```python
# I analize the partial correlation analysis

from statsmodel.tsa.stattools import acf # acf calculate the partial correlation analysis
nlags = 52*20 # 20 years, one year has 52 weeks
autocorr = acf(semanal.Tmean, nlags = nlags) # I want to analyze the outocorrelation of a series in nlags windows
```

```python
autocorr.shape
```

```python
# lets plot it
# we can see that the without lag correlation is exactly one
# also, we can see the correlation related with different lag values
# but after some lags, the correlation begin ro grow again
# each cicle has a duration of one year (52 weeks)
# in one year window we have the biggest correlation
# that means that the optimal look back in this series in one year
plt.figure(figsize=(8,5))
plt.scatter(np.arrange(nlags+1), autocorr)
plt.xlabel("lag")

# again, we are finding the size of the window that is necessary to define the size of the dataset
# I am looking for stacionality!
```

```python
# let plot the autocorrelarion with statsmodel library
# in this case we have our confidence interval (in blue)
# this confidence interval represents the area where the null H is fulfilled
# each correlation is asociated to a null hypothesis
# null H: cero correlation
# alternartive H: cero correlation (what I wan´t to reject), I wan´t to prove that correlation is not equal to cero.
# that means I am looking autocorrelations out of the confidence interval zone (not cero correlation)
from stats,models.graphics.tsaplots impor plot_acf
plot_acf(semanal.Tmean, lags=52*10)

```
